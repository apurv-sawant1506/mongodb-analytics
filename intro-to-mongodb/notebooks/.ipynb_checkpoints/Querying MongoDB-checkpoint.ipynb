```python
from pymongo import MongoClient
import pprint

client = MongoClient("mongodb://analytics:analytics-password@mflix-shard-00-00-fredv.mongodb.net:27017,mflix-shard-00-01-fredv.mongodb.net:27017,mflix-shard-00-02-fredv.mongodb.net:27017/test?ssl=true&replicaSet=mflix-shard-0&authSource=admin&retryWrites=true&w=majority")

print(client.mflix)
```

```python
db = client["mflix"]
collection = db["movies_initial"]
```

```python
movies = client.mflix.movies_initial
```

```python
pipeline1=[
    {
        '$group':{
            '_id': {'language':'$language'},
            'count': {'$sum':1}
                }
    },
    {
         '$sort' : {'count' : -1}
    }
]
```

```python
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline1)))
```

```python
pipeline2 = [
    {
    '$sortByCount' : "$language"
    }
]
```

```python
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline2)))
```

```python
pipeline3 = [
    {
      '$sortByCount' : '$language'  
    },
    
    {
        '$facet' : {
            'top language combinations' : [{'$limit' : 100}],
            'unusual combinations shared by' : [
                {'$skip' : 100} ,
                {
                    '$bucketAuto' : {
                        'groupBy' : '$count',
                        'buckets' : 5,
                        'output' : {
                            'language combinations' : {'$sum' : 1}
                        }
                    }
                }
            ]
        }
    }
]
```

```python
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline3)))
```

```python
pipeline4 = [
    {
        '$match' : {'language' : 'Korean, English'}
    }
]
```

```python
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline4)))
```

```python
filter = {'language' : 'Korean, English'}
```

```python
pprint.pprint(list(client.mflix.movies_initial.find(filter)))
```

```python
#This is how we can find the first 10 documents:
pipeline5=[{'$limit':10}]
for result in list(movies.aggregate(pipeline5)):
    pprint.pprint(result)
```
